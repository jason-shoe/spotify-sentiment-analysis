```python
playlists = pd.read_csv('listplaylist4.csv')
playlists
```

```python
def get_playlist_tracks(username,playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
def add_to_dataframe(playlist_id, playlist_name, playlist_mood, playlist_creator,songList,max_len):
    playlist_songs = get_playlist_tracks(playlist_creator,playlist_id)
    print("Songs: ",len(playlist_songs))
    for y in playlist_songs:
        try:
            a = sp.audio_features(y['track']['id'])[0]
            a.update(p_mood = playlist_mood)
            a.update(p_name = playlist_name)
            a.update(p_creator = playlist_creator)
            resultsSong = sp.audio_analysis(y['track']['id'])
            pitch = []
            timbre = []
            intensity = []
            rhythm = []
            for x in resultsSong['segments']:
                intensity.append(x['loudness_start'])
                for index in range(len(x['pitches'])):
                    pitch.append(x['pitches'][index])
                    timbre.append(x['timbre'][index])
                if(len(pitch) >= max_len):
                    break;
            for x in resultsSong['tatums']:
                rhythm.append(x['duration'])
                if(len(rhythm) >= max_len):
                    break;
            a.update(pitch_mean = statistics.mean(pitch))
            a.update(pitch_variance = statistics.variance(pitch))
            a.update(timbre_mean = statistics.mean(timbre))
            a.update(timbre_variance = statistics.variance(timbre))
            a.update(intensity_mean = statistics.mean(intensity))
            a.update(intensity_variance = statistics.variance(intensity))
            a.update(rhythm_mean = statistics.mean(rhythm))
            a.update(rhythm_variance = statistics.variance(rhythm))
            
            songList = songList.append(a, ignore_index=True)
            
            
            
        except:
            print("didnt work")
    return songList
```

```python
songList = pd.DataFrame()
added_uri = []
for index, row in playlists.iterrows():
    print(int(1000*(index+1)/len(playlists))/10)
    print(row['Playlist Name'])
    if(row['URI'][17:] not in added_uri):
        songList = add_to_dataframe(row['URI'][17:], row['Playlist Name'], row['Mood'], row['Creator'], songList, 500)
        added_uri.append(row['URI'][17:])
        print("Done")
    else:
        print("Already Done")
    print()
```
