```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials #To access authorised Spotify data
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import timeit
import statistics 
client_id = "dd555e5dcb514c99bd5674e226882fcf"
client_secret = "c185e52e743d4b6a9e14247aea41beb8"
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)

sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) #spotify object to access API
```

```python
def get_playlist_tracks(username,playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
def add_to_dataframe(playlist_id, playlist_name, playlist_mood, playlist_creator,songList):
    playlist_songs = get_playlist_tracks(playlist_creator,playlist_id)
    print("Songs: ",len(playlist_songs))
    for x in playlist_songs:
        try:
            a = sp.audio_features(x['track']['id'])[0]
            a.update(p_mood = playlist_mood)
            a.update(p_name = playlist_name)
            a.update(p_creator = playlist_creator)
            songList = songList.append(a, ignore_index=True)
        except:
            print("didnt work")
    return songList
```

```python
def add_to_dataframe_feature(playlist_id, playlist_name, playlist_mood, playlist_creator, songList, max_len):
    
    playlist_songs = get_playlist_tracks(playlist_creator,playlist_id)
    print("Songs: ",len(playlist_songs))
    for x in playlist_songs:
        try:
            resultsSong = sp.audio_analysis(x['track']['id'])
            pitch = []
            timbre = []
            intensity = []
            rhythm = []
            for x in resultsSong['segments']:
                intensity.append(x['loudness_start'])
                for index in range(len(x['pitches'])):
                    pitch.append(x['pitches'][index])
                    timbre.append(x['timbre'][index])
                if(len(pitch) >= max_len):
                    break;
            for x in resultsSong['tatums']:
                rhythm.append(x['duration'])
                if(len(rhythm) >= max_len):
                    break;
            songList = songList.append([playlist_mood, playlist_creator, playlist_name,
                                        statistics.mean(pitch), statistics.variance(pitch),
                                        statistics.mean(timbre), statistics.variance(pitch),
                                        statistics.mean(intensity), statistics.variance(intensity),
                                        statistics.mean(rhythm), statistics.variance(rhythm)], ignore_index=True)
        except:
            print("didnt work")
    return songList
```

```python
songListFeatures = pd.DataFrame(columns=['p_mood','p_creator','p_name','pitch_mean','pitch_variance','timbre_mean','timbre_variance','intensity_mean','intensity_variance','rhythm_mean','rhythm_'])
added_uri_features = []
```

```python
for index, row in playlists.iterrows():
    print(int(1000*(index+1)/len(playlists))/10)
    print(row['Playlist Name'])
    if(row['URI'][17:] not in added_uri_features):
        songListFeatures = add_to_dataframe_feature(row['URI'][17:], row['Playlist Name'], row['Mood'], row['Creator'], songListFeatures, 500)
        added_uri_features.append(row['URI'][17:])
        print("Done")
    else:
        print("Already Done")
    print()
```
